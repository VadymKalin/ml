Support Vector Machine (SVM) is a powerful supervised machine learning algorithm used for classification, regression, and outlier detection. The core idea of SVM is to find the optimal hyperplane that best separates data points of different classes in a high-dimensional space.

Here’s how it works in simple terms:
	•	SVM tries to maximize the margin between the two classes — this is the distance between the hyperplane and the nearest data points from each class (called support vectors).
	•	If the data is not linearly separable, SVM uses a technique called the kernel trick to transform it into a higher-dimensional space where a linear separator can be found.
	•	Common kernels include linear, polynomial, RBF (radial basis function), and sigmoid.

SVM is effective in high-dimensional spaces and works well when there is a clear margin of separation. However, it can be slower with large datasets and less effective when classes overlap significantly.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('income_evaluation.csv')
df.head()
```

```python
df.columns
```

```python
df[' income'].value_counts()
```

```python
col_names = df.columns
col_names = [v.strip() for v in col_names]
col_names
```

```python
df.columns = col_names
##df.drop(columns=['fnlwgt'], inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
bins = [16,24,64,90]
labels=['young','adult','old']
df['age_types'] = pd.cut(df['age'], bins=bins,labels=labels)
df['income_num'] = np.where(df['income'] == " >50K",1,0).astype('int16')
```

```python
df.head()
```

```python
df.loc[df['workclass']=='?', 'workclass']= np.NaN
df.loc[df['occupation']=='?', 'occupation']= np.NaN
df.loc[df['native-country']=='?', 'native_country']= np.NaN
```

```python
df = df.dropna(axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
def label_encoder(a):
    le = LabelEncoder()
    df[a] = le.fit_transform(df[a])
label_list = ['workclass', 'education','marital-status',
       'occupation', 'relationship', 'race', 'sex','native-country', 'income']
for i in label_list:
    label_encoder(i)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(df.drop(['income','age_types','income_num'],axis=1))

```

```python
scaled_features = scaler.transform(df.drop(['income','age_types','income_num'],axis=1))

columns=['age', 'workclass', 'education', 'education_num', 'marital_status',
       'occupation', 'relationship', 'race', 'sex', 'capital_gain',
       'capital_loss', 'hours_per_week', 'native_country']

df_scaled = pd.DataFrame(scaled_features,columns=columns)
df_scaled.head()

```

```python
from imblearn.combine import SMOTETomek
from imblearn.under_sampling import NearMiss

X = df_scaled
y= df.income

# Implementing Oversampling for Handling Imbalanced
smk = SMOTETomek(random_state=42)
X_res,y_res=smk.fit_resample(X,y)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res,y_res,test_size=0.20,random_state=101,shuffle=True)
from sklearn.model_selection import cross_val_score

from sklearn.svm import SVC
svc = SVC(random_state = 101)
accuracies = cross_val_score(svc, X_train, y_train, cv=5)
svc.fit(X_train,y_train)

print("Train Score:",np.mean(accuracies))
print("Test Score:",svc.score(X_test,y_test))
```
